```python
import numpy as np
import pandas as pd
import ast
import re
from itertools import chain

df_groups = pd.read_csv("data/groups_wMembers.csv")
df_groups["member"] = df_groups["member"].apply(ast.literal_eval)
```

```python
df_groups["cleaned_member"] = df_groups["member"].apply(
    lambda members: [member for member in members if member[0].isupper()]
)

df_groups["cleaned_member"] = df_groups["cleaned_member"].apply(
    lambda members: [member for member in members if member[-1].isalpha()]
)
```

```python
def remove_parentheses(name):
    return re.sub(r'\s*\(.*?\)', '', name).strip()

df_groups["cleaned_member"] = df_groups["cleaned_member"].apply(
    lambda members: [remove_parentheses(member) for member in members])
```

```python
musical_instruments = [
    'Sang', 'Vokal', 'Guitar', 'El-guitar', 'Piano', 'Keys', 'Klaver', 'Drums', 'Trommer', 'Bas', 'Bass', 'Violin', 'Saxophone', 'Saxofon' 'Trumpet',
    'Flute', 'Fløjte', 'Cello', 'Clarinet', 'Harpe', 'Trombone', 'Synthesizer', 'Keyboard', 'Sang/Guitar'
]

df_groups["cleaned_member"] = df_groups["cleaned_member"].apply(
    lambda members: [member.strip().split("-")[0] for member in members if member not in musical_instruments])
```

```python
all_members = list(chain.from_iterable(df_groups["cleaned_member"]))

print(all_members)
len(all_members)
```

```python
df_groups.to_csv("data/groups_wMembers.csv", index=False)
```
